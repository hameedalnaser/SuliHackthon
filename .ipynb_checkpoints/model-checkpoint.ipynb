```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('data/dataset.csv')
df
```

```python
#output
# 0 means very good soil for farming
# 1 good soil for farming
# 2 not that good for farming
df.tail(10)
```

```python
df['Output'].unique()
```

```python
x = df.drop('Output', axis=1)
```

```python
y=df['Output']
```

```python
#Transforming the data to make it between 0 and 1 so that the ML algorithm works better
scaler = MinMaxScaler(feature_range=(0, 1))
scaler = scaler.fit(x)
x = scaler.transform(x)
x
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(),annot=True)
```

```python
# features = df[['N', 'P', 'K', 'pH', 'EC', 'OC', 'S', 'Zn', 'Fe', 'Cu', 'Mn', 'B']]
features = df[['N', 'P', 'K', 'pH']]
target = df['Output']
#features = df[['temperature', 'humidity', 'ph', 'rainfall']]
labels = df['Output']
acc = []
model = []
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size = 0.2,random_state =2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn import tree
RF = RandomForestClassifier(n_estimators=20, random_state=0)
RF.fit(Xtrain,Ytrain)

predicted_values = RF.predict(Xtest)

x = metrics.accuracy_score(Ytest, predicted_values)
acc.append(x)
model.append('RF')
print("RF's Accuracy is: ", x)

print(classification_report(Ytest,predicted_values))
model
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(RF,features,target,cv=5)
score
```

```python
plt.figure(figsize=[10,1],dpi = 100)
plt.title('Accuracy Comparison')
plt.xlabel('Accuracy')
plt.ylabel('Algorithm')
sns.barplot(x = acc,y = model,palette='dark')
```

```python
accuracy_models = dict(zip(model, acc))
for k, v in accuracy_models.items():
    print (k, '-->', v)
```

```python
data = np.array([[138, 8.6, 560, 7.46]])
#                   0.62, 0.70, 5.90, 0.24, 0.31, 0.77, 8.71, 0.11]])
prediction = RF.predict(data)
print(prediction)
```
